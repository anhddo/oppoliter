```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def line_95_percent(cummulative_regret, label=None, color='b'):
    m = np.mean(cummulative_regret, axis=0)
    s = np.std(cummulative_regret, axis=0)
    u = m + s * 1.96
    l = m - s * 1.96
    plt.plot(m, label=label, color=color)
    plt.fill_between(range(cummulative_regret.shape[1]), u, l, alpha=0.2, color=color)
```

```python
m, n = 6, 10000
M = np.zeros((m, n))
x = range(n)
for i in range(m):
    reward, time_step = None, None
    with open('../tmp/avg-reward-step10e4-fourier4/result{}.pkl'.format(i), 'rb') as f:
        reward, time_step = pickle.load(f)

    y = np.interp(x,  time_step, reward)
    M[i] = y
```

```python
line_95_percent(M, 'avg')
plt.xlabel('')
plt.ylabel('Reward')
plt.ylim(0,200)
plt.show()
```
