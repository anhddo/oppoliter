```python
plt.rcParams['figure.figsize'] = [7, 5]
plt.rcParams.update({'font.size': 13})
df_opt = pd.read_csv('opt_100_150000')
df_greedy = pd.read_csv('greedy_100_150000')
df_noex = pd.read_csv('noex_100_150000')
lw = 2.5
ax = df_opt['reward'].cumsum()[::1000].plot(linewidth=lw)
df_greedy['reward'].cumsum()[::1000].plot(linewidth=lw)
df_noex['reward'].cumsum()[::1000].plot(linewidth=lw)
ax.legend(['Optimistic VI', r'$\epsilon$-greedy', 'Standard VI'])
ax.set_ylabel('Cummulative reward')
ax.set_xlabel(r'$t \times 10^4$')
# ax.set_xticklabels(range(0,16,5), range(0, 150000, 5000))
plt.xticks(range(0, 160000, 30000), range(0,16,3))
ax.set_title('N-chain, $N=100$')
plt.savefig('../../notebook/n_chain.pdf')
```
